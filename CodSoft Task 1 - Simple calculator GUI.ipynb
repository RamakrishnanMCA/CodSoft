```python
# Design a simple calculator with basic arithmetic operations.Prompt the user to input two numbers and an operation choice.
# Perform the calculation and display the result.

from tkinter import *

content=""

def onclick(get):
    global content
    content = content + str(get)
    text.set(content)
    
def answer():
    try:
        global content
        tot=str(eval(content))
        text.set(tot)
        content=""
        print(tot)
    
    except:
        text.set("error")
        content=""
        print("error")

def clear():
    global content
    content =""
    text.set("")
    
if __name__ == "__main__":  
    
    win=Tk()
    win.title("Simple Calculator")
    text=StringVar()

    l1 = Label(win,text="Calculator",fg="Gold",font=(0,25))
    l1.grid(row=0,column=0,columnspan=4,padx=10,pady=10)
    
    e1=Entry(win,textvariable=text,width=35)
    e1.grid(row=1,column=0,columnspan=4,padx=10,pady=10)

    bt1=Button(win,text='1',padx=40,pady=20,activebackground="Aqua",command=lambda: onclick(1))
    bt2=Button(win,text='2',padx=40,pady=20,activebackground="Aqua",command=lambda: onclick(2))
    bt3=Button(win,text='3',padx=40,pady=20,activebackground="Aqua",command=lambda: onclick(3))
    bt4=Button(win,text='4',padx=40,pady=20,activebackground="Aqua",command=lambda: onclick(4))
    bt5=Button(win,text='5',padx=40,pady=20,activebackground="Aqua",command=lambda: onclick(5))
    bt6=Button(win,text='6',padx=40,pady=20,activebackground="Aqua",command=lambda: onclick(6))
    bt7=Button(win,text='7',padx=40,pady=20,activebackground="Aqua",command=lambda: onclick(7))
    bt8=Button(win,text='8',padx=40,pady=20,activebackground="Aqua",command=lambda: onclick(8))
    bt9=Button(win,text='9',padx=40,pady=20,activebackground="Aqua",command=lambda: onclick(9))
    bt0=Button(win,text='0',padx=40,pady=20,activebackground="Aqua",command=lambda: onclick(0))
    
    btans=Button(win,text='=',padx=40,pady=25,activebackground="DarkCyan",command=answer)
    btclear=Button(win,text='clear',padx=130,pady=25,activebackground="DeepSkyBlue",command=clear)
    btadd=Button(win,text='+',padx=40,pady=25,activebackground="DeepSkyBlue",command=lambda: onclick("+"))
    btsub=Button(win,text='-',padx=40,pady=25,activebackground="DeepSkyBlue",command=lambda: onclick("-"))
    btmul=Button(win,text=' * ',padx=40,pady=25,activebackground="DeepSkyBlue",command=lambda: onclick("*"))
    btdiv=Button(win,text='/',padx=40,pady=25,activebackground="DeepSkyBlue",command=lambda: onclick("/"))


    bt1.grid(row=2,column=0)
    bt2.grid(row=2,column=1)
    bt3.grid(row=2,column=2)
    bt4.grid(row=3,column=0)
    bt5.grid(row=3,column=1)
    bt6.grid(row=3,column=2)
    bt7.grid(row=4,column=0)
    bt8.grid(row=4,column=1)
    bt9.grid(row=4,column=2)
    bt0.grid(row=5,column=0)
    
    btans.grid(row=5,column=1)
    btclear.grid(row=7,column=0,columnspan=4)
    btadd.grid(row=5,column=2)
    btsub.grid(row=6,column=0)
    btmul.grid(row=6,column=1)
    btdiv.grid(row=6,column=2)
    
  
    win.mainloop()
```
